```python
clustring_data =  merged_data.groupby('LYLTY_CARD_NBR').agg({
'LYLTY_CARD_NBR':'count',
'PROD_QTY':'sum',
'TOT_SALES':['sum']
}).reset_index()
clustring_data.columns = ['consumer_id','Frequancy','Quantity','Total_Payed']
clustring_data
```

```python

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

```python
data = clustring_data.iloc[:,1:]
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
```

```python
wcss = []
for i in range(1,10) :
    km = KMeans(n_clusters=i,init='k-means++',max_iter=1000,random_state=0)
    km.fit(data_scaled)
    wcss.append((i,km.inertia_))

```

```python
wcss = pd.DataFrame(wcss,columns=['K','inertia'])
```

```python
sns.lineplot(wcss,x='K',y='inertia')
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
x_pca = pca.fit_transform(data-data_scaled)
```

```python
explained_variance = pca.explained_variance_ratio_
print(f'Explained variance for each component: {explained_variance}')
```

```python
km = KMeans(n_clusters=3,init='random',max_iter=1000)
y_kmeans = km.fit_predict(x_pca)
```

```python
clustring_data['Clusters'] = y_kmeans
clustring_data
```

```python
sns.scatterplot(x=x_pca[:,0],y=x_pca[:,1],hue=clustring_data['Clusters'],palette='muted')
```

```python
# Get cluster centers
centroids = km.cluster_centers_

# Print the coordinates of the centroids
print("Cluster Centers (Centroids):\n", centroids)
```

```python

analysis = clustring_data.groupby('Clusters').agg({
    'consumer_id':'count', 'Frequancy':['mean','sum','std'], 'Quantity':['mean','sum','std'], 'Total_Payed':['mean','sum','std'] })
analysis
```
